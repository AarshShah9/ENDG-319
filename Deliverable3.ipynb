```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
a = pd.read_excel("/content/drive/MyDrive/Maths.xlsx")
a.head()
```

```python


df = pd.DataFrame(a)
df = a.loc[:, ['higher', 'schoolsup', 'famsup', 'Medu', 'Fedu',  'G3']].copy()
df.rename(columns={'higher': 'higher_class'}, inplace=True)


df['higher_class'].replace('yes', 1, inplace=True)
df['higher_class'].replace('no', 0, inplace=True)

df['schoolsup'].replace('yes', 1, inplace=True)
df['schoolsup'].replace('no', 0, inplace=True)

df['famsup'].replace('yes', 1, inplace=True)
df['famsup'].replace('no', 0, inplace=True)


df.head()
```

```python

final_grade_0 = df["G3"].loc[df["higher_class"] == 0]
final_grade_1 = df["G3"].loc[df["higher_class"] == 1]

mother_edu_0 = df["Medu"].loc[df["higher_class"] == 0]
mother_edu_1 = df["Medu"].loc[df["higher_class"] == 1]

father_edu_0 = df["Fedu"].loc[df["higher_class"] == 0]
father_edu_1 = df["Fedu"].loc[df["higher_class"] == 1]

schoolsup_0 = df["schoolsup"].loc[df["higher_class"] == 0]
schoolsup_1 = df["schoolsup"].loc[df["higher_class"] == 1]

famsup_0 = df["famsup"].loc[df["higher_class"] == 0]
famsup_1 = df["famsup"].loc[df["higher_class"] == 1]



fig,axs = plt.subplots(figsize=(15, 3))

axs1 = plt.subplot2grid(shape=(1, 4), loc = (0, 0))
axs2 = plt.subplot2grid(shape=(1, 4), loc = (0, 1))
axs3 = plt.subplot2grid(shape=(1, 4), loc = (0, 2))

plt.tight_layout()


axs1.hist(final_grade_0, edgecolor='b', fc='none', label="No Higher Education")
axs1.hist(final_grade_1, edgecolor='r', fc='none', label="Seeking Higher Education")
axs1.set_xlabel("Final Grade")
axs1.set_ylabel("Frequency")
axs1.legend()



axs2.scatter(mother_edu_0, final_grade_0, c="blue", label="No Higher Education")
axs2.scatter(mother_edu_1, final_grade_1, c="orange", s=6, label="Seeking Higher Education")
axs2.set_xlabel("Mother Education")
axs2.set_ylabel("Final Grade")
axs2.legend()

axs3.scatter(father_edu_0, final_grade_0, c="blue", label="No Higher Education")
axs3.scatter(father_edu_1, final_grade_1, c="orange", s=6, label="Seeking Higher Education")
axs3.set_xlabel("Father Education")
axs3.set_ylabel("Final Grade")
axs3.legend()


fig,axs = plt.subplots(figsize=(15, 3))

axs4 = plt.subplot2grid(shape=(1, 4), loc = (0, 0))
axs5 = plt.subplot2grid(shape=(1, 4), loc = (0, 1))

plt.tight_layout()


axs4.scatter(schoolsup_0, final_grade_0, c="blue", label="No Higher Education")
axs4.scatter(schoolsup_1, final_grade_1, c="orange", s=6, label="Seeking Higher Education")
axs4.set_xlabel("School Support")
axs4.set_ylabel("Final Grade")
axs4.legend()


axs5.scatter(famsup_0, final_grade_0, c="blue", label="No Higher Education")
axs5.scatter(famsup_1, final_grade_1, c="orange", s=6, label="Seeking Higher Education")
axs5.set_xlabel("Family Support")
axs5.set_ylabel("Final Grade")
axs5.legend()
```
