### Task 1

```python
from sklearn.datasets import load_breast_cancer
import pandas as pd

# a)
cancer_data = load_breast_cancer(return_X_y=True, as_frame = True)
a = cancer_data[0]
a["typeofcancer"] = cancer_data[1]

a.shape
```

```python
# b)
df = a.iloc[:, [0, 2, 3, -1]]

# i)
print(df)

# ii)
print(df[17:22])
```

### Task 2

```python
import matplotlib.pyplot as plt


mean_perimeter_0 = df["mean perimeter"].loc[df["typeofcancer"] == 0]
mean_radius_0 = df["mean radius"].loc[df["typeofcancer"] == 0]
mean_area_0 = df["mean area"].loc[df["typeofcancer"] == 0]


mean_perimeter_1 = df["mean perimeter"].loc[df["typeofcancer"] == 1]
mean_radius_1 = df["mean radius"].loc[df["typeofcancer"] == 1]
mean_area_1 = df["mean area"].loc[df["typeofcancer"] == 1]

fig,axs = plt.subplots(figsize=(15, 3))

axs1 = plt.subplot2grid(shape=(1, 4), loc = (0, 0))
axs2 = plt.subplot2grid(shape=(1, 4), loc = (0, 1))
axs3 = plt.subplot2grid(shape=(1, 4), loc = (0, 2))
plt.tight_layout()




axs1.hist(mean_radius_0, edgecolor='b', fc='none', label="c0(Malignant)")
axs1.hist(mean_radius_1, edgecolor='r', fc='none', label="c1(Benign)")
axs1.set_xlabel("mean radius")
axs1.set_ylabel("Frequency")
axs1.legend()



axs2.scatter(mean_perimeter_0, mean_radius_0, c="blue", label="c0(M)")
axs2.scatter(mean_perimeter_1, mean_radius_1, c="orange", label="c1(B)")
axs2.set_xlabel("mean perimeter")
axs2.set_ylabel("mean radius")
axs2.legend()


axs3.scatter(mean_area_0, mean_radius_0, c="blue", label="c0(M)")
axs3.scatter(mean_area_1, mean_radius_1, c="orange", label="c1(B)")
axs3.set_xlabel("mean area")
axs3.set_ylabel("mean radius")
axs3.legend()



```

```python
mean_concavity_0 = a["mean concavity"].loc[a["typeofcancer"] == 0]
mean_radius_0 = df["mean radius"].loc[a["typeofcancer"] == 0]
mean_concavep_0 = a["mean concave points"].loc[df["typeofcancer"] == 0]
mean_symmetry_0 = a["mean symmetry"].loc[a["typeofcancer"] == 0]

mean_concavity_1 = a["mean concavity"].loc[a["typeofcancer"] == 1]
mean_radius_1 = df["mean radius"].loc[df["typeofcancer"] == 1]
mean_concavep_1 = a["mean concave points"].loc[a["typeofcancer"] == 1]
mean_symmetry_1 = a["mean symmetry"].loc[a["typeofcancer"] == 1]

fig,axs = plt.subplots(figsize=(15, 3))

axs1 = plt.subplot2grid(shape=(1, 4), loc = (0, 0))
axs2 = plt.subplot2grid(shape=(1, 4), loc = (0, 1))
axs3 = plt.subplot2grid(shape=(1, 4), loc = (0, 2))
axs4 = plt.subplot2grid(shape=(1, 4), loc = (0, 3))
plt.tight_layout()

axs1.hist(mean_radius_0, edgecolor='b', fc='none', label="c0(Malignant)")
axs1.hist(mean_radius_1, edgecolor='r', fc='none', label="c1(Benign)")
axs1.set_xlabel("mean radius")
axs1.set_ylabel("Frequency")
axs1.legend()

axs2.scatter(mean_concavity_0, mean_radius_0, c="blue", label="c0(M)")
axs2.scatter(mean_concavity_1, mean_radius_1, c="orange", label="c1(B)")
axs2.set_xlabel("mean concavity")
axs2.set_ylabel("mean radius")
axs2.legend()

axs3.scatter(mean_concavep_0, mean_radius_0, c="blue", label="c0(M)")
axs3.scatter(mean_concavep_1, mean_radius_1, c="orange", label="c1(B)")
axs3.set_xlabel("mean concave points")
axs3.set_ylabel("mean radius")
axs3.legend()

axs4.scatter(mean_symmetry_0, mean_radius_0, c="blue", label="c0(M)")
axs4.scatter(mean_symmetry_1, mean_radius_1, c="orange", label="c1(B)")
axs4.set_xlabel("mean symmetry")
axs4.set_ylabel("mean radius")
axs4.legend()
```
